## Creating dataframe with necessary columns only

```python
df1 = df[['Year', 'Month', 'Day', 'Country', 'State', 'Region', 'city', 'latitude', 'longitude', 'Attack_Type', 'Killed',
         'Wounded', 'Group', 'Target', 'Target_Type', 'Weapon_type']]
df1.head(10)
```

## Univariate Analysis
**The term univariate analysis refers to the analysis of one variable, prefix "uni" means "one".The purpose of univariate analysis is to understand the distribution of the values for a single variable.**

```python
df1['Attack_Type'].value_counts()
```

```python
(df1['Attack_Type'].value_counts()/df1.shape[0])*100
```

```python
df1['Attack_Type'].value_counts().plot(kind='pie',title = 'Percentage of Attack Type Distribution', autopct='%1.2f%%', figsize=(15,15))
```

**Conclusion :** Highest number of attacks happend with Bombing/Explosion (48.57%) & Second highest by Armed Assaults (23.48%)

```python
plt.figure(figsize=(10,10))
sns.barplot(df1['Country'].value_counts()[:10].index,df1['Country'].value_counts()[:10].values, palette = 'magma')
plt.style.use('ggplot')
plt.title('Top 10 Countries affected with Terrorism',fontsize=20)
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

## Bivariate Analysis

```python
# Region with most wounded people 
plt.subplot(1,2,1)
sns.barplot(df1.groupby(['Region'])['Wounded'].sum().sort_values(ascending=False).index,
            df1.groupby(['Region'])['Wounded'].sum().sort_values(ascending=False).values,
           palette = 'bone')
plt.title('Region wise people Wounded', fontsize=20)
plt.xlabel('Region', fontsize=20)
plt.ylabel('People Wounded', fontsize=20)
plt.xticks(rotation=90)


# Region with most killed people
plt.subplot(1,2,2)
sns.barplot(df1.groupby(['Region'])['Killed'].sum().sort_values(ascending=False).index,
            df1.groupby(['Region'])['Killed'].sum().sort_values(ascending=False).values,
           palette = 'bone')
plt.title('Region wise people Killed', fontsize=20)
plt.xlabel('Region', fontsize=20)
plt.ylabel('People Killed', fontsize=20)
plt.xticks(rotation=90)

plt.gcf().set_size_inches(15,7)
```

```python
# Region with most wounded people 
plt.subplot(1,2,1)
sns.barplot(df1.groupby(['Country'])['Wounded'].sum().sort_values(ascending=False).head(10).index,
            df1.groupby(['Country'])['Wounded'].sum().sort_values(ascending=False).head(10).values,
           palette = 'hot')
plt.title('Country wise people Wounded', fontsize=20)
plt.xlabel('Country', fontsize=20)
plt.ylabel('People Wounded', fontsize=20)
plt.xticks(rotation=90)


# Region with most killed people
plt.subplot(1,2,2)
sns.barplot(df1.groupby(['Country'])['Killed'].sum().sort_values(ascending=False).head(10).index,
            df1.groupby(['Country'])['Killed'].sum().sort_values(ascending=False).head(10).values,
           palette = 'hot')
plt.title('Country wise people Killed', fontsize=20)
plt.xlabel('Country', fontsize=20)
plt.ylabel('People Killed', fontsize=20)
plt.xticks(rotation=90)

plt.gcf().set_size_inches(15,7)
```

## Conclusion
**After performing the Exploratory Data Analysis we get the following insights from the data:**

- Private Citizens and Property were attacked most followed by Military, Police, Government and so on.
- Middle East & North Africa was most affected among the top affected region. Most of the people in this region were either wounded or killed.
- Iraq was the country which was most affected by terror attacks and had maximum number of killed and wounded people.
- The State and City that was most affected was Baghdad
- In the last decade 2014 had most number of terror attacks. That was around 16500 attacks during this year which means on average 45 attacks per day.
- The most common attack type was Bombing/Explosion.
- Explosives have been consistently the most popular weapon of choice for terrorists.
- Taliban became more active since 2012 and they are responsible for the most of the terror attacks
