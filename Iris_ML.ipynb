### Machine Learning
### Dataset Link : http://archive.ics.uci.edu/ml/datasets/Iris

### The aim is to predict the species of the flower (setosa, versicolor or virginica) based on the lengths of their sepal and petal's length and width.

The iris data set has 150 rows with 3 classes (species of iris flower) of 50 instances each.

The Datasource is iris.data which is in the CSV format (comma seperated variables). 
It contains 5 columns and 150 rows.

#### IMPORTING LIBRARIES

```python
# IMPORTING ALL THE NECESSARY LIBRARIES FOR OUR PROJECT

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

#### LOADING THE DATASET  

```python
#naming the columns
columns = ['Sepal_length','Sepal_width','Petal_length','Petal_width','Class_labels']

#loading the data
df=pd.read_csv('iris.data', names=columns)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
# Visualising the dataset
sns.pairplot(df, hue='Class_labels')
```

#### SEPERATING INPUT AND OUTPUT COLUMNS 

```python
data = df.values

X = data[:,0:4]
Y = data[:, 4]
```

#### SPLITTING DATA INTO TRAINING AND TESTING 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)

# testsize=0.2 indicates that 0.2*total data size will be our testing size and rest all will be training.
```

### MODEL1 : SUPPORT VECTOR MACHINE ALGORITHM

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train, Y_train)
```

```python
prediction_1 = model_svc.predict(X_test)

#accuracy
from sklearn.metrics import accuracy_score
print('The accuracy of the Model1 is ', accuracy_score(Y_test, prediction_1)*100,'%')
```

### MODEL2 : LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(X_train, Y_train)
```

```python
prediction_2 = model_LR.predict(X_test)

#accuracy
from sklearn.metrics import accuracy_score
print('The accuracy of the Model2 is ', accuracy_score(Y_test, prediction_2)*100,'%')
```

### MODEL3 : DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC = DecisionTreeClassifier()
model_DTC.fit(X_train, Y_train)
```

```python
prediction_3 = model_DTC.predict(X_test)

#accuracy
from sklearn.metrics import accuracy_score
print('The accuracy of the Model3 is ', accuracy_score(Y_test, prediction_3)*100,'%')
```

```python
#New Inputs
New_X = np.array([[4,2,1,3],[4,4,4,4],[4.3,2.4,3,3]])

prediction_new = model_LR.predict(New_X)
print(f"Prediction of Species {prediction_new}")
```
